# Analysis of Survey Data Professionals

## Introduction

- **Where is this data from ?**<br>
This data is collected from data profesionals by youtube channel name Alex The Analyst by posting a survey to his linkedin, twitter and other social media platforms. 630 data professionals responded to the survey. It measures the satisfaction of data professionals in their current job and whether they got into the industry by changing their career or not and how difficult it is to get into. It also contain an information about the respondent country, job title, gender ,salary, education level and many other information.

- **Who is this analysis for ?** <br>
Im hoping that this analysis will help as a point of reference for  anyone who is interested in data science field and want to know about the satisfaction of data professionals in their current job and how difficult it is to get into the industry. If you are a data professional and want to know how your satisfaction in your current job compare to other data professionals, this analysis will help you to get some insights.

## Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv('Cleaned BI.csv')
df.head()
```

```python
df = df.dropna(axis=1, how='all')
df.head()
```

```python
df.columns
```

```python
jobTitleDistribution = df['Q1 - Which Title Best Fits your Current Role?'].value_counts()
carreerDistribution = df['Q2 - Did you switch careers into Data?'].value_counts()
```

```python
groupedData = df.groupby(['Q1 - Which Title Best Fits your Current Role?','Q2 - Did you switch careers into Data?']).size().unstack()
groupedData
```

## Data Visualization

### Job and Salary Distribution

```python
groupedData.plot(kind='bar', stacked=True)
plt.title('Job Title Distribution')
plt.ylabel('Number of People')
plt.xlabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
carreerDistribution = df['Q2 - Did you switch careers into Data?'].value_counts()
print(carreerDistribution)
```

```python
career_switchers = df[df['Q2 - Did you switch careers into Data?'] == 'Yes']
career_switchers_salary = career_switchers['Q3 - Current Yearly Salary (in USD)']
career_switchers_salary.describe()
```

```python
non_career_switchers = df[df['Q2 - Did you switch careers into Data?'] == 'No']
non_career_switchers_salary = non_career_switchers['Q3 - Current Yearly Salary (in USD)']
non_career_switchers_salary.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(career_switchers_salary, label='Career Switchers', fill=True)
sns.kdeplot(non_career_switchers_salary, label='Non Career Switchers', fill=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Based on the salary survey data the salary difference are not significant for those who switch into data professionals carrer. This mean, you can switch into data professionals without worrying about the salary being different for those with data professionals background.

```python
career_switchers_difficulty = career_switchers['Q7 - How difficult was it for you to break into Data?'].value_counts()
career_switchers_difficulty
```

```python
color_map = {
    'Very Easy': 'lightgreen',
    'Easy': 'green',
    'Neither easy nor difficult': 'gray',
    'Difficult': 'orange',
    'Very Difficult': 'red'
}

colors = [color_map[category] for category in career_switchers_difficulty.index]

plt.figure(figsize=(10, 6))
plt.pie(career_switchers_difficulty, labels=career_switchers_difficulty.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Difficulty to Break into Data')
plt.show()
```

Based on the breakdown, most people does not find it difficult to switch into data professionals. This is a good news for those who are considering to switch into data professionals career. Although, there are some people who find it difficult to switch into data professionals, but the percentage shows that most people are able to change without any problem.

```python
job_title_salary = df.groupby('Q1 - Which Title Best Fits your Current Role?')['Q3 - Current Yearly Salary (in USD)'].mean()
job_title_salary = job_title_salary.sort_values(ascending=False)
job_title_salary

plt.figure(figsize=(10, 6))
sns.barplot(x=job_title_salary.index, y=job_title_salary.values, hue=job_title_salary.index, palette='viridis', legend=False)
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Data scientist have the highest salary among the data professionals even though data analysis is more common roles in the industry.

```python
education_salary = df.groupby('Q12 - Highest Level of Education')['Q3 - Current Yearly Salary (in USD)'].mean()
education_salary = education_salary.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=education_salary.index, y=education_salary.values, hue=education_salary.index, palette='viridis', legend=False)
plt.title('Average Salary by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```

chances are, you will get more salary if you have a higher education level. the difference between those who have a degree in Phd and those who have a degree below that is significant. But rest assured, if you're trying to get into data professionals without a degree, you can still get a salary without to many differences or equal to those who have a degree.

```python
industry_job_title = df.groupby(['Q4 - What Industry do you work in?','Q1 - Which Title Best Fits your Current Role?']).size().unstack()
industry_job_title

plt.figure(figsize=(10, 6))
sns.heatmap(industry_job_title, cmap='RdYlGn_r')
plt.title('Job Distribution on Industries')
plt.xlabel('Job Title')
plt.ylabel('Industry')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
industry_salary = df.groupby('Q4 - What Industry do you work in?')['Q3 - Current Yearly Salary (in USD)'].mean()
industry_salary = industry_salary.sort_values(ascending=False)
industry_salary

plt.figure(figsize=(10, 6))
sns.barplot(x=industry_salary.values, y=industry_salary.index, hue=industry_salary.index, palette='viridis', legend=False)
plt.title('Average Salary by Industry')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.show()
```

Among all of the industries that the survey takers worked in. Retail has the most average salary for data professionals. if you are considering to switch into data professionals with a high salary, you might want to consider working in retail industry but be mindfull of the salary and work life balance satisfaction.

```python
country_salary = df.groupby('Q11 - Which Country do you live in?')['Q3 - Current Yearly Salary (in USD)'].mean()
country_salary = country_salary.sort_values(ascending=False)
country_salary

plt.figure(figsize=(10, 6))
sns.barplot(x=country_salary.values, y=country_salary.index, hue=country_salary.index, palette='viridis', dodge=False, legend=False)
plt.title('Average Salary by Country')
plt.xlabel('Average Salary')
plt.ylabel('Country')
plt.show()
```

United States have the highs salary for data professionals. But keep in mind, that higher salary is connected to the living cost of that certain country, so low salary on a certain country might be enough to live a good life. Though it is a good reference if you want to get a high salary by doing a remote work or moving to another country. For that, US, Canada, and Australia are the 3 best options

### Salary and Work/Life Balance Satisfaction

```python
salary_satisfaction = df['Q6 - How Happy are you in your Current Position with the following? (Salary)'].mean()
work_life_balance_satisfaction = df['Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)'].mean()

print('Salary Satisfaction:', salary_satisfaction)
print('Work Life Balance Satisfaction:', work_life_balance_satisfaction)
```

```python
industry_salary_satisfaction = df.groupby('Q4 - What Industry do you work in?')['Q6 - How Happy are you in your Current Position with the following? (Salary)'].mean()
industry_salary_satisfaction = industry_salary_satisfaction.sort_values(ascending=False)
industry_salary_satisfaction

plt.figure(figsize=(10, 6))
sns.barplot(x=industry_salary_satisfaction.values, y=industry_salary_satisfaction.index, hue=industry_salary_satisfaction.index, palette='viridis', dodge=False, legend=False)
plt.title('Salary Satisfaction by Industry')
plt.xlabel('Salary Satisfaction (out of 10)')
plt.ylabel('Industry')
plt.xlim(0, 10)
plt.show()
```

```python
industry_workLifeBalance_satisfaction = df.groupby('Q4 - What Industry do you work in?')['Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)'].mean()
industry_workLifeBalance_satisfaction = industry_workLifeBalance_satisfaction.sort_values(ascending=False)
industry_workLifeBalance_satisfaction

plt.figure(figsize=(10, 6))
sns.barplot(x=industry_workLifeBalance_satisfaction.values, y=industry_workLifeBalance_satisfaction.index, hue=industry_workLifeBalance_satisfaction.index, palette='viridis', dodge=False, legend=False)
plt.title('Work/Life Balance Satisfaction by Industry')
plt.xlabel('Work/Life Balance Satisfaction (out of 10)')
plt.ylabel('Industry')
plt.xlim(0, 10)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Q6 - How Happy are you in your Current Position with the following? (Salary)'], label='Salary', fill=True)
sns.kdeplot(df['Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)'], label='Work/Life Balance', fill=True)
plt.title('Salary and Work/Life Balance Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Density')
plt.legend()
plt.show()
```

People who work in Logistics industy have the highest work/life balance satisfaction while construction are the lowest among all of them expect for the unemployed. Although logistic industry have the highest salary and worklife balance, it is still averaging in 6 to 7 out of 10. The satisfaction for data profesional in general for those who fill in the survey is not that high. This is something to consider. It is fine to assume that the salary and work ratio is not very well balanced. 

### Job Preference and Motivation

```python
programming_language = df['Q5 - Favorite Programming Language'].value_counts()
programming_language
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=programming_language.index, y=programming_language.values, hue=programming_language.index, palette='viridis', legend=False)
plt.title('Favorite Programming Language Distribution')
plt.xlabel('Number of People')
plt.ylabel('Programming Language')
plt.show()
```

- Data Professional prefer to use python than any other language. If you are considering to switch into data professionals, you might want to learn python first. It is the most used language among data professionals.
- Though, its does not mean that you should not learn other language. Some people consider that SQL is not a programming language, but it still have a strong ties with managing data.
- While DAX is the lowest among all of the language, it might be the same case as SQL where it is not considered as a programming language but you will frequently use it if you work in excel or other BI tools.

```python
job_search_priority = df['Q8 - If you were to look for a new job today, what would be the most important thing to you?'].value_counts()
job_search_priority
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=job_search_priority.index, y=job_search_priority.values, hue=job_search_priority.index, palette='viridis', legend=False)
plt.title('Job Search Priority')
plt.xlabel('Number of People')
plt.ylabel('Priority')
plt.xticks(rotation=45, ha='right')
plt.show()
```

It is obvious that most people are motivated by the salary, but its still good to know the order of priority for people where they want to change jobs. Most people will still choose salary over Remote work and good work life balance.
